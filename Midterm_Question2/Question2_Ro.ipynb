```python
import sys, os, requests, json, re  
import matplotlib.pyplot as plt

from collections import Counter
from pylab import *

cwd = os.getcwd()

os.chdir(cwd)
months = [i for i in range(1,2)]
year = 2016

for j in months:
    url = 'http://api.nytimes.com/svc/archive/v1/'+str(year)+'/'+str(j)+'.json?api-key=2e34eb9b843f4d47adb836f2072600c4'
    resp = requests.get(url)
    data = json.loads(resp.text)
    mon = '0'+str(j) if j < 10 else str(j)
    with open('archive_'+str(year)+'_'+mon+'.json','w') as fp:
        json.dump(data,fp)
```

```python
os.chdir(cwd)
os.chdir(cwd)
from glob import glob
import os
maindir = 'R:\\Rohini Midterm\\Ques2\\Data\\'
os.chdir(maindir)
cwd = os.getcwd()

dt = []
wordCount = []
for each in glob('*'):
    json_data=open(each).read()
    data = json.loads(json_data)
    sectionName = []
    typeOfMaterial = []
    source = []
    for i in range(len(data["response"]["docs"])):
        temp = data["response"]["docs"][i]["section_name"]
        sectionName.append(str(temp))
        temp = data["response"]["docs"][i]["type_of_material"]
        typeOfMaterial.append(str(temp.encode('ascii','ignore').decode('ascii')))
        temp = data["response"]["docs"][i]["source"]
        source.append(str(temp))
        wordCount.append(temp)

    counter1 = Counter(sectionName)
    names1 = counter1.keys()
    counts1 = counter1.values()

    indexes = arange(len(names1))
    width = 0.7
    plt.bar(indexes, counts1, width)
    plt.xticks(indexes - width * 2, names1, rotation=90)
    plt.show()
```
